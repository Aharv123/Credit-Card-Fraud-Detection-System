```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None) 
pd.set_option('display.expand_frame_repr', False) 

df = pd.read_csv("creditcard.csv")
df.head(50)
```

```python
df.describe()      # here we describe the whole dataset with count, means std, min and quantile ranges
```

```python

df.shape      # to show the shape of the dataset
```

```python
df.info()
```

```python
df.isnull().sum()       # shows 0 to all so clean data no Null values. We are good to go
```

```python

# 0 = Legit, 1 = Fraud
print("\nClass Distribution:")
print(df['Class'].value_counts())
print("\nFraudulent Transaction Percentage: {:.4f}%".format(df['Class'].mean() * 100))

#CountPlot visualizations here
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, hue='Class', palette='Set2', legend=False)
plt.title('Class Distribution (0 = Legit, 1 = Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['Amount'], bins=100, color='orange')
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.xlim(0, 3000) 
plt.show()

# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='Amount', data=df, hue='Class', palette='coolwarm', legend=False)
plt.title("Transaction Amount by Class")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Amount")
plt.yscale('log')  
plt.show()

# Histogram with respect to Time destribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Time'], bins=100, color='teal')
plt.title("Transaction Time Distribution (seconds since first transaction)")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()

# Heatmap for top correlated features
correlation_matrix = df.corr()
top_corr = correlation_matrix['Class'].abs().sort_values(ascending=False).head(11).index

plt.figure(figsize=(10, 8))
sns.heatmap(df[top_corr].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Top Correlated Features with Class")
plt.show()
```
